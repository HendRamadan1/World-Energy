```python
import pandas as pd

def NullPercent(df):
    """
    Returns a DataFrame showing missing value count and percentage for each column.
    Only includes columns with at least one missing value.
    Ensures full display of all rows.
    """
    # Calculate missing values per column
    missing = df.isnull().sum().sort_values(ascending=False)
    
    # Filter only columns with missing values
    missing = missing[missing > 0]
    
    # If no missing values, return empty DataFrame
    if missing.empty:
        return pd.DataFrame(columns=['Missing Count', 'Missing Percent (%)'])
    
    # Compute percentage relative to total rows
    percent_missing = (missing / len(df)) * 100
    
    # Create result DataFrame
    missing_df = pd.DataFrame({
        'Missing Count': missing,
        'Missing Percent (%)': percent_missing.round(2)
    })
    
    return missing_df
```

```python
def getMissingValues(df):
    missing_values = df.isnull().sum().sort_values(ascending=False)
    missing_values = missing_values[missing_values > 0]
    missing_values = missing_values / len(df)
    return [missing_values], missing_values.__len__()
```
